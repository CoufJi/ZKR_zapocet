# Vygenèrova šifra

Jedná se o symetrickou polyalfabetickou substituční šifru.

Vychází z šifry Leona Battisty Albertiho z roku 1467, kdy byla uveřejněna. Zajímavostí je, že finální podobu jí dal až Giovan Battista Bellaso v roce 1553.

## Důvod vzniku
Důvodem k vytvoření této šifry byla reakce na snadnou prolomitelnost tehdějších metod pomocí frekvenční analýzy, například Ceaserova šifra, ta využívá jen jednu abecedu po celou šifru, cílem bylo vytvořit polyalfabetickou šifru, která by:
- odolala útokům na základě frekvence písmen
- měnila šifrovací abecedu v průběhu textu
- skryla stejná písmena o.t. pod různá písmena v z.š 

## Pravidla použití 
Je to polyalfabetická šifra, k šifrovaní se používá celý klíč, které se dokola opakuje pro šifrovaní o.t. Písmenka v klíči určují, o kolik se posouvá písmenko o.t., takže například:
- klíč: b, o.t.: c -> 2 (b) * 3 (c) = 6 (f)

## Důsledky porušení kryptografického protokolu
Důležité je použít klíč, ale takový klíč, ve kterém se moc neopakují písmenka a není příliš krátký, protože pokud je klíč krátký a jednoduchý, v š.t. se vzor se začne ihned opakovat.

Na druhou stranu by byl zase problém, pokud by klíč byl až příliš dlouhý a náhodný.

## Prolomení
Šifra byla prolomena v 19. století pruským důstojníkem **Friedrichem Kasiskim** pomocí Kasiskiho testu:
1. V zašifrovaném textu hledá opakující se shluky písmen - vzdálenost mezi shluky pomůže najít společného dělitele a odhalit délku klíče
2. Když se zná délka klíče, š.t. se dle délky klíče rozdělí na daný počet sloupců a pomocí frekvenční analýzy rozluští š.t. (nejčastější se hledají písmenka jako "E" a "A")
